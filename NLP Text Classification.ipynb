```python
def features(name) :
    # convert name into lower case.
    name = name.lower()
    feature_dict = {}       # Empty Dictionary

    # Getting length of 'names'
    feature_dict['Length']= len(name)
    feature_dict['first_letter']= name[0]
    feature_dict['last_letter']= name[-1]

    # Alphabet , vowel count
    feature_dict['vowel_count'] = 0
    for i in string.ascii_lowercase:             # i = a,b,c,.......,z
        feature_dict[f' count_{i}'] = name.count(i)
        

    # vowel count
    if (i in 'aeiou') and ( i in name):
        feature_dict['Vowel_count'] = feature_dict['Vowel_count'] + name.counr(i)

    return feature_dict

```
